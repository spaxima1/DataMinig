```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Stars.csv")
```

```python
data    
```

```python
color=data['Color']
lbl = LabelEncoder()
a=lbl.fit_transform(color)
data['Color']=a
```

```python
data
```

```python
Spectral_Class=data['Spectral_Class']
lbl = LabelEncoder()
a=lbl.fit_transform(Spectral_Class)
data['Spectral_Class']=a
```

```python
data['newLType']=data['newL'].copy()
```

```python
data['Type'].unique()
```

```python
import matplotlib.pyplot as plt
```

```python
for i in data['Type'].unique():
    plt.subplot()
    data['newLType'].loc[data['Type']==i].plot(kind = "kde")
    plt.show()
```

```python
listtemp=[]
for i in data['Type'].unique():
    index=data['newLType'].loc[data['Type']==i].copy()
    index=index.replace(np.nan,index.median())
    listtemp+=list(index.values)

data['newLType']=listtemp
```

```python
data[['L','newL','newLmedian','newLType']].loc[data['newL']!=data['L']]
```

```python
data['newTempmedianType']=data['newTemp'].copy()
for i in data['Type'].unique():
    plt.subplot()
    data['newTempmedianType'].loc[data['Type']==i].plot(kind = "hist")
    plt.show()
```

```python
listtemp=[]
for i in data['Type'].unique():
    index=data['newTempmedianType'].loc[data['Type']==i].copy()
    index=index.replace(np.nan,index.median())
    listtemp+=list(index.values)

data['newTempmedianType']=listtemp
```

```python
data[['Temperature','newTemp','newTempmedian','newTempmedianType']].loc[data['newTemp']!=data['Temperature']]
```
